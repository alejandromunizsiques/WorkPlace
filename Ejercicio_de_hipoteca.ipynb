#Preguntas:#

¿Cuánto sería el pago mensual para una hipoteca de 30 años (30 * 12 = 360 meses totales) con un capital de 400.000 dólares y una TAE de 3% (tasa de interés con honorarios incluidos)? 

```python


P = int(input("Ingrese cantidad del prestamo"))
r = float(input("Ingrese tasa de interes"))
n = int(input("Ingrese cantidad de cuotas (meses) "))

dividendo = r*(1+r)**n
divisor = (1+r)**n-1
Pago_mensual = P*(dividendo/divisor)

print(Pago_mensual)


```
